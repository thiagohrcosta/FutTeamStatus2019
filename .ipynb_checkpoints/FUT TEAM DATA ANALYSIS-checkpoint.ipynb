```python
# Creating a DataFrame with two diferent Leagues using the conditional OR ( " | ") 
# to select players from Calcio Serie A or League 1

df5 = pd.read_csv("myFutTeam.csv")
cond7 = df5["League"] == "Serie A"
cond8 = df5["League"] == "League 1"
df5[cond7 | cond8]
```

```python
# Creating a DataFrame with two diferent Leagues using the conditional AND
# to select players from Calcio Serie A to select players above with Overall >= 90

df6 = pd.read_csv("myFutTeam.csv")
cond9 = df6["League"] == "Serie A"
cond11 = df6["Overall"] >= 90
df6[cond9 & cond11]
```

```python
# Creating a DataFrame with two diferent Leagues using the conditional OR and AND
# to select players from Calcio Serie A or League 1, who has Overall >= 90

df7 = pd.read_csv("myFutTeam.csv")
cond12 = df7["League"] == "Serie A"
cond13 = df7["League"] == "League 1"
cond14 = df7["Overall"] >= 90
df7[(cond12 | cond13) & cond14]
```

```python
# Here we use the function "Between" to set two values to search "Price In Coins" between
# players who cost >= 150000 and <= 200000

df10 = pd.read_csv("myFutTeam.csv")
df10[df10["Price In Coins"].between(150000, 200000)]
```

```python
# Same function that above, but searching for players that have overall between 89 and 90.

df10[df10["Overall"].between(89, 90)]
```

```python
def overallPlayers(Overall):
    if Overall < 85:
        return "Weak Player"
    elif Overall > 85 and Overall <= 91:
        return "Good Player"
    else:
        return "Excellent Player +92"
```

```python
df9["Overall"].apply(overallPlayers).head()
```

```python
df9["Overall"].apply(overallPlayers).value_counts()
```

```python
#plt.style.use("ggplot")
df9["Overall"].apply(overallPlayers).value_counts().plot(kind = "barh")
```

```python
def leaguePlayers(League):
    if League == "League 1":
        return "League 1 Players - FRANCE"
    elif League == "Serie A":
        return "Serie A Players - ITALY"
    elif League == "Premier League":
        return "Premier League Players - ENGLAND"
    else:
        return "Other Leagues"
```

```python
df9["League"].apply(leaguePlayers).head()
```

```python
df9["League"].apply(leaguePlayers).value_counts().plot(kind = "barh")
```
